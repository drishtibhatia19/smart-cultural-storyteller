## Problem Statement

Most existing digital storytelling platforms generate generic narratives
that lack cultural depth, emotional alignment, and personalization.

Traditional systems:
- Do not adapt stories based on cultural context
- Ignore emotional tone and moral perspective
- Produce single, unstructured narratives
- Offer limited user control over story direction

personally meaningful, culturally grounded, or emotionally engaging.

There is a need for an intelligent storytelling system that can:
- Preserve cultural narratives
- Adapt stories based on user-selected emotions and morals
- Generate structured, multi-scene stories
- Support multimedia extensions such as images and narration

```python
payload = {
    "culture": "Ancient India",
    "mode": "cultural",
    "emotion": "Hope",
    "moral": "True strength lies in compassion",
    "user_input": "A story about devotion and courage"
}

payload
```

```python
def build_story_prompt(payload):
    return f"""
You are a master storyteller rooted in {payload['culture']} traditions.

Story mode: {payload['mode']}
Emotional tone: {payload['emotion']}
Moral lesson: {payload['moral']}

User idea:
{payload['user_input']}

Generate a story divided into multiple scenes.
Each scene must include:
- title
- detailed narrative text
"""
```

```python
for i, scene in enumerate(story_output["scenes"], start=1):
    print(f"Scene {i}: {scene['title']}")
    print(scene["text"])
    print("-" * 40)
```
